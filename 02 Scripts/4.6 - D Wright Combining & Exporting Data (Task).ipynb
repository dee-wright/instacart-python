#### 4.6.3 - Importing Pickle

```python
df_ords_prods = pd.read_pickle(os.path.join(path, '02 Data', 'Cleaned', 'orders_products_combined.pkl'))
```

#### 4.6.4 - Checking Imported Dataframe

```python
df_ords_prods.shape
```

#### 4.6.5 - Merging df_prods & df_ords_prods

```python
# Merging dataframes on product_id column
orders_products_merged = df_ords_prods.merge(df_prods, on='product_id')
```

```python
orders_products_merged.head()
```

```python
# Checking merge rates - inner
orders_products_merged['_merge'].value_counts()
```

```python
orders_products_merged_outer = df_ords_prods.merge(df_prods, on='product_id', how='outer')
```

```python
# Checking merge rates - outer
orders_products_merged_outer['_merge'].value_counts()
```

#### 4.6.7 - Exporting new dataframe

```python
# Removing _merge flag for tidiness
orders_products_merged.drop(columns='_merge')
```

```python
# Exporting merged dataframe as pickle due to size
orders_products_merged.to_pickle(os.path.join(path, '02 Data', 'Cleaned', 'orders_products_merged.pkl'))
```
